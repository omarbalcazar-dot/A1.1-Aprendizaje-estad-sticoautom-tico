En esta celda puse la libreria pandas, que hace trabajar con datos en forma de tablas.

Se carga el archivo Obesidad.csv usando read_csv().

Use head(10) para ver las primeras 10 filas.

```python
import pandas as pd
df = pd.read_csv("Obesidad.csv")
df.head(10)
```

Si estuviera en el equipo de investigacion, mediria la actividad fisica semanal (horas de ejercicio a la semana). Seria una variable cuantitativa, porque expresa como un n√∫mero que indica horas de ejercicio.

En esta celda escogi la variable de NivelDeObesidad como salida, ya que es el objetivo principal del estudio.
Como es una variable cualitativa, uso value_counts() para contar cuantas personas pertenecen a cada categoria.

```python
sal = df["NivelDeObesidad"]
sal.value_counts()
```

En esta celda hice una grafica de dispersion de la poblacion masculina.
El eje X representa el peso y el eje Y el nivel de obesidad.

Utilice el alpha para hacer los puntos mas transparentes, para que sea mas facil de visualizar.

```python
%matplotlib inline

import matplotlib.pyplot as plt

df_fem = df[df["Sexo"] == "Male"]

plt.scatter(df_fem["Peso"], df_fem["NivelDeObesidad"], alpha= 0.1)
plt.xlabel("Peso (kg)")
plt.ylabel("Nivel de Obesidad")
plt.title("Relacion Peso - Nivel de Obesidad (Masculino)")
plt.show()
```
